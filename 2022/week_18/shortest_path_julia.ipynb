```julia
c = fill(Inf, (7, 7))
c[1, 2], c[1, 3], c[1, 4] = 1, 5, 3
c[2, 4], c[2, 5] = 9, 6
c[3, 6] = 2
c[4, 6] = 4
c[5, 7] = 4
c[6, 7] = 1
c[7, 7] = 0
c
```

```julia
function T(q)
    Tq = similar(q)
    n = length(q)
    for x in 1:n
        Tq[x] = minimum(c[x, :] + q[:])
    end
    return Tq
end
```

```julia
using PyPlot

fig, ax = plt.subplots()

n = 7
q = zeros(n)
ax.plot(q)

for i in 1:3
    new_q = T(q)
    ax.plot(1:n, new_q)
    q = new_q
end
```
